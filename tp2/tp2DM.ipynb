```python
import matplotlib.pyplot as plt

def create_scatter_plot(x_values, y_values, title, x_label, y_label):
    plt.figure(figsize=(10, 6))
    plt.scatter(x_values, y_values, color='blue', alpha=0.6)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

```python
def create_histogram(values, title, x_label, y_label, number_of_bins):
    plt.figure(figsize=(10, 6))
    plt.hist(values, bins=number_of_bins, color='green', alpha=0.7)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

```python
import seaborn as sns

def create_box_plot(values, title, y_label, show_outliers=True):
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=values, showfliers=show_outliers)
    plt.title(title)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

## 1. Scatter Plot (Diagramme de dispersion):
- Un scatter plot montre la relation entre deux variables continues.
- detection de groupes et des patterns
- detection des outliers
